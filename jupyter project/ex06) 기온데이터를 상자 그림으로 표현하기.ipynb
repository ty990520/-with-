# 기온데이터를 상자 그림으로 표현하기

```python
# 1~1000까지의 임의의 수를 추출해 상자그림 그리기
import matplotlib.pyplot as plt
import random
result=[]

for i in range(13):
    result.append(random.randint(1,1000))
    
print(sorted(result))

plt.boxplot(result)
plt.show()

# 위치값 알아내기
import numpy as np
result=np.array(result)
print("1/4: ", str(np.percentile(result,25)))
print("2/4: ", str(np.percentile(result,50)))
print("3/4: ", str(np.percentile(result,75)))
```

```python
import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
result=[]

for row in data:
    if row[-1]!='':
        result.append(float(row[-1]))
        
import matplotlib.pyplot as plt
plt.boxplot(result)
plt.show()
```

```python
import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
aug=[]
jan=[]

for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
        if month=='08':
             aug.append(float(row[-1]))
        if month=='01':
             jan.append(float(row[-1]))
       
        
import matplotlib.pyplot as plt
#plt.boxplot(aug)
#plt.boxplot(jan)
plt.boxplot([aug,jan]) #분리됨
plt.show()
```

```python
#최고 기온 데이터를 월별로 구분하여 표현하기
import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
month_list=[],[],[],[],[],[],[],[],[],[],[],[]

for row in data:
    month=row[0].split('-')[1]
    if row[-1]!='':
        month_list[int(month)-1].append(float(row[-1]))
            
import matplotlib.pyplot as plt
plt.boxplot(month_list)
plt.show()
```

```python
#8월 일별 기온 데이터를 상자 그림으로 표현하기
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)

day=[]
for i in range(31):
    day.append([])
    
for row in data:
    date_split=row[0].split('-')
    if row[-1]!='':
        if date_split[1]=='08':
            day[int(date_split[2])-1].append(float(row[-1]))
  
plt.style.use('ggplot') #그래프의 스타일을 지정
plt.figure(figsize=(10,5),dpi=300)
plt.boxplot(day, showfliers=False) #아웃라이어(이상치) 값 생략
 
plt.show()
```

```python
# 1년 중 일교차(daily temperture range)가 가장 큰 달은 몇 월일까?
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)

#대상 정하기 : 2019년도를 대상

month=[]
for i in range(12):
    month.append([])
    
for row in data:
    date_split=row[0].split('-')
    if date_split[0]=='2019':
        if row[-1]!='' and row[-2]!='':
            daily_temp= float(row[-1])-float(row[-2]) #일교차
            month[int(date_split[1])-1].append(round(daily_temp,1))
            #print(date_split[1],date_split[2],row[-1],row[-2],round(daily_temp,1))
plt.style.use('ggplot') #그래프의 스타일을 지정
plt.figure(figsize=(10,5),dpi=300)
plt.boxplot(month, showfliers=False) #아웃라이어(이상치) 값 생략
plt.show()
#해석: 2019년도에 일교차가 가장 큰 달은 4월, 일반적으로 일교차가 높은 날은 5월(중위수)
```

```python
# 1년 중 일교차(daily temperture range)가 가장 큰 달은 몇 월일까?
import csv
import matplotlib.pyplot as plt
f=open('seoul.csv')
data=csv.reader(f)
next(data)

#대상 정하기 : 1907~2020
month=[]
for i in range(12):
    month.append([])
    
for row in data:
    date_split=row[0].split('-')
    if row[-1]!='' and row[-2]!='':
        daily_temp= float(row[-1])-float(row[-2]) #일교차
        month[int(date_split[1])-1].append(round(daily_temp,1))
        
plt.style.use('ggplot') #그래프의 스타일을 지정
plt.figure(figsize=(10,5),dpi=300)
plt.boxplot(month, showfliers=False) #아웃라이어(이상치) 값 생략
plt.show()
#해석: 1907~2020년도에 일교차가 가장 큰 달은 4월, 일반적으로 일교차가 높은 날은 5월(중위수)
```
