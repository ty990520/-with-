```python
import csv
import matplotlib.pyplot as plt
f=open('age.csv')
data=csv.reader(f)
next(data)
result=[]

for row in data:
    if '연수구' in row[0] and '송도' in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))#콤마를 빈문자로 바꾸기
            
#result=list(map(int,result))
#result=[int(x) for x in result]
print(result)
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.title('인천 연수구 송도 지역의 인구 구조')
plt.plot(result)
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt
f=open('age.csv')
data=csv.reader(f)
next(data)
result=[]
name1=input('인구 구조가 알고 싶은 지역의 행정구역(시) :')
name2=input('군/구:')
name3=input('읍/면/동(제외):')

for row in data:
    if name1 in row[0] and name2 in row[0] and name3 in row[0] :
        for i in row[3:]:
            result.append(int(i.replace(',','')))
            
print(result)

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.title(name1+" "+name2+" "+name3+' 지역의 인구 구조')
plt.plot(result)
plt.show()
```
