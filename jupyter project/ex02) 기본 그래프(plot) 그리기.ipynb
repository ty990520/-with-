# 기본 그래프 그리기

### 기본 그래프 세 단계
1) 라이브러리 불러오기
2) plot()함수에 데이터 입력하기
3) show()로 그래프 보여주기

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([10,20,30,40]) #y축 값으로 입력됨
plt.show()
```

```python
plt.title('plotting')
plt.plot([1,2,3,4],[12,43,25,15]) #x축, y축
plt.show()
# x축 데이터와 y축 데이터의 개수가 맞지 않으면 에러 발생
```

```python
plt.title('legend')
plt.plot([10,20,30,40],label='asc')
plt.plot([40,30,20,10],label='desc')
plt.legend(loc=1) #legend(): 범례 표시 #loc=범례위치
plt.show()
```

```python
plt.title('color & linestyle')
plt.plot([10,20,30,40],color='skyblue', linestyle='--',label='skyblue / dashed')
plt.plot([40,30,20,10],color='pink', ls=':', label='pink / dotted')
plt.legend(loc=2)
plt.show()
```

```python
plt.title('marker')
plt.plot([10,20,30,40],'r.',label='circle')
plt.plot([40,30,20,10],'g^',label='triangle up')
plt.legend()
plt.show()
```
