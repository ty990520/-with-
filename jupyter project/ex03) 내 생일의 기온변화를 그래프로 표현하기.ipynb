```python
# 1999.05.20~2020.05.20 기간 동안의 최고기온과 최저기온 표현
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv')
data=csv.reader(f)
next(data)
high=[] # 최고기온 값을 저장할 리스트
low=[] # 최저기온 값을 저장할 리스트


for row in data:
    if row[-1]!='' and row[-2]!='':
        date = row[0].split('-')
        if int(date[0])>=1999:
            if date[1]=='05' and date[2]=='20':
                high.append(float(row[-1]))
                low.append(float(row[-2]))

plt.rc('font',family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False #마이너스 기호 깨짐 방지 (어차피 5월은 영하기온 없어서 ㄱㅊ)
plt.figure(figsize=(8,5))
plt.title('내 생일의 기온 변화 그래프')
plt.plot(high,'r',label='high')
plt.plot(low,'b',label='low')
plt.legend()
plt.show()
```

```python
# 평균 기온이 상승하는 그래프 표현 ★모르겠음★
import csv
import matplotlib.pyplot as plt

f= open('seoul.csv')
data=csv.reader(f)
next(data)
mean=[]
for row in data:
    if row[-1]!='' and row[-2]!='':
        month = row[0].split('-')[1]
        if int(month)==5:
             mean.append((float(row[-1])+float(row[-2]))/2)
plt.figure(figsize=(10,5))
plt.plot(mean,'b')
plt.show()
```
